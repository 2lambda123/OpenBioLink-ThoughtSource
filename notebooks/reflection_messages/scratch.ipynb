```python
import openai

openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "The Los Angeles Dodgers won the World Series in 2020."},
        {"role": "user", "content": "Where was it played?"}
    ]
)
```

```python
#Use code to add in models

#Remove CoTs 

import json

with open("./experiment_data/med_qa_500_first_cohere_gpt-3.5-turbo.json", 'r') as f:
    data = json.load(f)


#data = Collection.from_json("./med_qa_first_gpt-3.5-turbo_gpt-4.json")
# Loop through each test item
for test_item in data['med_qa']['test']:
    # Remove the second element of 'generated_cot' list
    
    str_dict = test_item['generated_cot'][1]['model']
    # Convert the string to a valid JSON format by replacing single quotes with double quotes
    str_dict = str_dict.replace("'", '"')

    # Convert the string to a dictionary
    dict_data = json.loads(str_dict)

    # Update the 'name' value
    dict_data['name'] = "gpt-3.5-turbo"

    # Convert the dictionary back to a string, if needed
    str_dict = json.dumps(dict_data)
    test_item['generated_cot'][1]['model'] = str_dict



# Now you can convert the data back to a JSON string if needed
# modified_json_str = json.dumps(data)

with open("./experiment_data_with_models/med_qa_500_first_cohere_gpt-3.5-turbo.json", 'w') as json_file:
    json.dump(data, json_file)







```
