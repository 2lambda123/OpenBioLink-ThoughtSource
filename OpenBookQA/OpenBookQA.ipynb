```python
import json_lines
import json
```

```python
input_address = '4-OpenBookQA\\dev_complete.jsonl'
output_address = '4-OpenBookQA\\dev-new.json'

instruction = 'Answer the question by selecting the correct choice: '

samples_list = []

with open(input_address) as input_file:
    for line in json_lines.reader(input_file):
        print(line)
        
        question = line['question']
        question_stem = question['stem']
        print('\nQuestion:', question_stem)
        
        choices = []
        choices_list = question['choices']
        for ch in choices_list:
            choice = {
                "choice_text" : ch['text'],
                "choice_label" : ch['label']
            }
            choices.append(choice)
        
        print('\nChoices:', choices)
        
        problem = {
            "question" : question_stem,
            "choices" : choices
        }
        
        chain_of_thought = line['fact1'] + '.'
        print('\nCOT:', chain_of_thought)
        answer = line['answerKey']
        print('\nAnswer:', answer)
        
        
        sample = {
            "instruction" : instruction,
            "problem" : problem,
            "cot" : chain_of_thought,
            "answer" : answer
        }
        samples_list.append(sample)
        
        print('-' * 100)
        

json_object = json.dumps(samples_list, indent = 4)
#print(samples_list)
with open(output_address, "w") as output_file:
    output_file.write(json_object)
```
