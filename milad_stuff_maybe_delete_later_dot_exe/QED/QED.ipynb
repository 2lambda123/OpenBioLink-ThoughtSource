```python
import json_lines
import json
```

```python
input_address = '3-QED\\qed-dev.jsonl'
output_address = '3-QED\\test-new.json'

instruction = 'Regarding the paragraph, answer the question.'

samples_list = []

with open(input_address) as input_file:
    for line in json_lines.reader(input_file):
        #print(line)
        
        question = line['question_text'] + '?'
        print('Question:', question)
        
        #---------- These information is used for COT as explanation
        annotation = line['annotation']
        print('annotation:', annotation)
        
        if (annotation['explanation_type'] in ('none', 'multi_sentence')):
            print('-' * 100)
            continue
        
        ref_eq = annotation['referential_equalities']
        q_ref_text = ''
        s_ref_text = ''
        for i in ref_eq:
            q_ref = i['question_reference']
            q_ref_text += q_ref['string']
            s_ref = i['sentence_reference']
            s_ref_text += s_ref['string']
        
        answer = annotation['answer']
        answer_s_ref_text = ''
        answer_p_ref_text = ''
        for j in answer:
            answer_s_ref = j['sentence_reference']
            answer_s_ref_text += answer_s_ref['string']
            answer_p_ref = j['paragraph_reference']
            answer_p_ref_text += answer_p_ref['string']
        
        selected_sentence = annotation['selected_sentence']
        
        
        chain_of_thought = {
            "referential_equalities" :{
                "question_reference" : q_ref_text,
                "sentence_reference" : s_ref_text
            },
            "answer":{
                "sentence_reference" : answer_s_ref_text,
                "paragraph_reference" : answer_p_ref_text
            },
            "selected_sentence" : selected_sentence['string']
        }
        
        
        #----------
        
        paragraph_text = line['paragraph_text']
        print('\nParagraph Text:', paragraph_text)
        
        problem = {
            "paragraph" : paragraph_text,
            "question" : question
        }
        
        answer = []
        answer_list = line["original_nq_answers"]
        print('\nAnswer:', answer_list)
        if (len(answer_list) > 0):
            answers = answer_list[0]
            for answer_dict in answers:
                answer.append(answer_dict['string'])
        print('\nAnswer:', answer)
        
        
        
        
        sample = {
            "instruction" : instruction,
            "problem" : problem,
            "cot" : chain_of_thought,
            "answer" : answer
        }
        samples_list.append(sample)
        
        print('-' * 100)
        

print('\nSAMPLES CONVERTED:', len(samples_list))        

json_object = json.dumps(samples_list, indent = 4)
#print(samples_list)
with open(output_address, "w") as output_file:
    output_file.write(json_object)
```
